# Индивидуальное $№2$.

## Есть кривая, заданная параметрически с помощью двух функций $x(t)$ и $y(t)$, где $t$ - параметр, изменяющийся в интервале $[a, b]$. Наша задача состоит в вычислении длины этой кривой на заданном интервале.

```python
# импортируем библиотеки
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x) + 1  # Пример функции y = f(x)
```

```python
def calculate_volume(a, b):
    x = np.linspace(a, b, 100)
    y = f(x)
    volume = np.pi * np.trapz(y ** 2, x)
    return volume
```

```python
# Заданные параметры
a = 0  # Левая граница
b = np.pi  # Правая граница
```

```python
# Расчет объема
volume = calculate_volume(a, b)
# Вывод результата
print("Объем тела:", volume)
```

```python
# Построение графика функции и тела
x = np.linspace(a, b, 100)
y = f(x)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, color='gray', alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Объем тела, полученного вращением кривой y = f(x) вокруг оси x')
plt.grid(True)
plt.show()
```
